```python
#1 #2
import pandas as pd 
import numpy as np 

h = pd.read_csv('./house_pricing.csv'); 
for x in h:
    print(x + ' : ' + str(h[x].dtypes))
```

```python
#3
import matplotlib.pyplot as pyplot

pyplot.scatter( h['SalePrice'], h['LotArea'],c = 'red')
pyplot.ylabel('LotArea')
pyplot.xlabel('SalePrice')
pyplot.show()
```

```python
#4
d = h[(h['SalePrice'] < 500000) & (h['LotArea'] < 20000) ]

pyplot.scatter(d['LotArea'], d['SalePrice'],c = 'red')
pyplot.xlabel('LotArea')
pyplot.ylabel('SalePrice')

pyplot.show()
```

```python
#5 #6
d = h[(h['LotFrontage'] <= 200) & (h['LotArea'] <= 100000) ]

pyplot.scatter(d['LotFrontage'], d['LotArea'],c = 'red')
pyplot.ylabel('LotArea')
pyplot.xlabel('LotFrontage')

pyplot.show()
```
