# Étude des données de vente d'un site web.

Le fichier se compose d'un export des ventes de produits du site web Anglais sur l'année 2011 (on imagine qu'on est l'année +1)

Notre but est de comprendre le flux client sur le site afin de se développer à l'internationnal.

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

h = pd.read_csv('./dataset.csv'); 
h.InvoiceDate = pd.to_datetime(h.InvoiceDate)
h = h.sort_values(by='InvoiceDate')
h.head(10)
```

```python
h.describe()
```

Il semble qu'il y ait des données abérantes, certains prix étant en dessous de 0. Il faut les supprimer.

```python
h = h[(h['UnitPrice'] > 0) & (h['Quantity'] > 0)]
```

## Nous allons déjà regarder quand le site fait le plus de vente

```python
#nb commande / mois

import statistics as stat
from collections import OrderedDict
import datetime

# quantié / mois

d = {}
s = {}

for (i,x) in h.groupby(['InvoiceNo']):
    date = x.iloc[0]['InvoiceDate'].strftime("%m/%Y")
    if date in d:
        d[date] = d[date] + 1
    else :
        d[date] = 1
med =  stat.median(d.values())   

for x in d.keys():
    s[x] = med

d = OrderedDict(sorted(d.items(), key=lambda t: datetime.datetime.strptime(t[0], '%m/%Y')))
s = OrderedDict(sorted(s.items(), key=lambda t: datetime.datetime.strptime(t[0], '%m/%Y')))
fig_dims = (15, 5)
fig, ax = plt.subplots(figsize=fig_dims)        
sns.lineplot(x=d.keys(), y=d.values(), ax=ax)
sns.lineplot(x=s.keys(), y=s.values(), ax=ax)
```

On peut constater que le nombre median de vente depuis mars 2011 est d'environ 1600 commandes / mois, sauf sur le mois de novembre où on observe un pic de commandes. 

On utilise ici une médianne et non un moyenne qui serait faussée par le pic du mois de novembre et ne serait donc pas représentatif du nombre de ventes mensuel moyen.

## Nous allons maintenant chercher à étudier les utilisateurs

```python
#utilisateurs / pays

d = pd.DataFrame({'Customer': [], 'Country': []})
total = len(h)

for (i,x) in h.groupby(['Country']):
    if len(x)/total > 0.01:
        d = d.append({'Customer': len(x), 'Country': i}, ignore_index=True)

fig_dims = (15, 5)
fig, ax = plt.subplots(figsize=fig_dims)      

ax = sns.barplot(x='Country', y='Customer', data=d, ax=ax)
```

On peut déjà constater que la très grande majorité des clients viennent du Royaume-Unis. Le site ne semble donc que très peu développé à l'internationnal

```python
#panier moyen

d = {}
s = {}

for (i,x) in h.groupby(['InvoiceNo']):
    date = x.iloc[0]['InvoiceDate'].strftime("%m/%Y")
    if not date in d:
        d[date] = {}
    else :
        m = x['UnitPrice'] * x['Quantity']
        d[date].update({i: m.sum()})


r = {}
for (i,x) in d.items():
    r[i] = stat.median(x.values())

med =  stat.median(r.values())   

for x in d.keys():
    s[x] = med
    
fig_dims = (15, 5)
fig, ax = plt.subplots(figsize=fig_dims)        

sns.lineplot(x=r.keys(), y=r.values(), ax=ax)
sns.lineplot(x=s.keys(), y=s.values(), ax=ax)
```

On peut voir ci-dessus le panier moyen par utilisateur chaque mois. On constate que le panier moyen se trouve autour de 240€. On observe toutefois un pic au mois de septembre mais pas au mois de novembre qui a lui enregistré le plus fort taux de commandes.

```python
#nb commande / pays / mois

d = pd.DataFrame()
r = h[h['Country'] != 'United Kingdom']

r['InvoiceDate'] = r['InvoiceDate'].dt.strftime('%m/%Y')

for (i,x) in r.groupby(['InvoiceDate']):
    date = x.iloc[0]['InvoiceDate']
    for (c, row) in x.groupby(['Country']):
        d = d.append({date: stat.mean(row.Quantity), 'Country': c}, ignore_index=True)

d = d.groupby(['Country']).sum()
d.replace(0, 'NaN').dropna()

plt.figure(figsize=(16,9))
sns.heatmap(d, linewidths=1, cmap="Blues")
```

Le graphique est relativement intéressant dans la mesure où il nous indique pour les pays autres que le Royaume-Unis la quantité de produits commandés par mois.
On ne prend pas le Royaume-Unis dans ce graphique car il fausse les données en raison du grand nombre de commandes pour ce pays. Cela nous permet d'observer pour chaque pays le pic de popularité. Cette information peut être intéressante afin d'optimiser le site pour un événement particulier dans un pays.

```python
#commande / sexe
d = pd.DataFrame()

for (i, x) in h.groupby('CustomerId'):
    size = len(x.groupby('InvoiceNo'))
    if size < 10:
        d = d.append({'Sexe': x.iloc[0].Sexe, 'Commands': size}, ignore_index=True)

plt.figure(figsize=(16,9))
sns.boxplot(data=d, y='Sexe',x='Commands')
        
```

Le graphique ci-dessus, nous permet de voir la répartition homme / femme en fonction du nombre de commandes. Nous ne prenons que les clients ayants commandés moins de 10 fois car ils représentent plus de 75% de notre échantillion et les autres valeurs fausses notre graphique. 

Ce dernier nous permet de constater que sur notre échantillion les femmes commandent dans 50% des cas entre 2 et 5 fois alors que ce n'est qu'entre 1 et 5 pour les hommes. On peut donc supposer ques les femmes de cet échantillion reviennent plus sur le site que les hommes.

## On peut maintenant étudier les produits et leur popularité

```python
#prix unitaire / quantité

fig_dims = (15, 5)
fig, ax = plt.subplots(figsize=fig_dims)      
sns.scatterplot(y='UnitPrice', x='Quantity', data=h[(h['UnitPrice'] < 10) & (h['Quantity'] < 200)], ax=ax)
```

Ce graphique nous permet de voir la répartition de la quantité commandé en fonction du prix unitaire. Nous ne prenons ici que les commandes avec un produit ayant un prix unitaire inférieur à 10 car c'est plus de 75% de notre échantillon. Et ça nous permet d'avoir une meilleure vue de la répartition.

```python
fig_dims = (8, 15)
fig, ax = plt.subplots(figsize=fig_dims)      

sns.barplot(x='Quantity', y='Country', data=h.sort_values(['Country', 'Quantity']), ax=ax)
```

Ce graphique nous permet de constater la quantité moyenne commandée par produit par pays. 
On constate que le Royaume-unis, bien qu'étant le pays ou il y a le plus d'utilisateurs est aussi celui ou les utilisateurs commandes le moins en quantité. Il faut toutefois préciser que certains pays on peut de commandes mais avec beaucoup de quantité.

### Conclusion

Notre analyse nous permet de dégager 2 principaux axes de développement pour l'activité du site.

- Il semble être nécessaire de développer le site à l'étranger afin d'accroître le trafique. Ce développement pourrait notamment passer par la personnalisation du site en fonction des besoins utilisateurs par pays et des produits qu'ils consomment le plus.
- Les femmes semblent commander plus de fois sur le site que les hommes. Notre échantillion de données ne permet pas de définir pourquoi. Il serait intéressant de disposer de plus d'informations sur les produits achetés par chaque type afin de comprendre quelle partie est à développer pour fidéliser plus de clients
